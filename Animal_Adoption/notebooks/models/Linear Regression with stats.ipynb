```python
# Define the features and target variable
features = ['age_upon_intake_(years)']
target = 'time_in_shelter_days'
```

```python
import matplotlib.pyplot as plt

# Plot!
plt.scatter(df['age_upon_intake_(years)'], df['time_in_shelter_days'], alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('age_upon_intake_days')
plt.ylabel('time_in_shelter_days')
plt.show()
```

```python
# Training the model (statsmodels)
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
# Ensure that the shapes are correct
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Fit the Linear Regression model using Ordinary Least Squares regression 
model = sm.OLS(y_train, X_train).fit() # fit the model
```

```python
print(model.summary()) # print the summary of the model
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluate the model using metrics
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE: ', mse)
print('R-squared: ', r2)
```
